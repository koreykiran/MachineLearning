# Project - Bring ML to your area of interest. 

As we briefly talked about in the first day of class, the project is playing with the problem of **your** interest. 
Find a problem that you and your teammates like. It can be any subject!
Find data related to the problem or simulation model to play with. 
If you do not know where to start, take a look at our [resource page](https://webpages.uncc.edu/mlee173/teach/itcs6156/resources.html) for problems/applications/data. 

The project requires a certain level of difficulty, so you can have a valuable experience. Your final report should demonstrate what you have learned, especially something new about machine learning. Please consult with me for your proposal. 

As the course is in a graduate level, the work is expected to be good preliminary work that you can publish after more investigation. 
Since this is a team project, you need to state each member's contribution in **both** proposal and final report.

The recommended structure of the report contains:

- Title
- Name(s)
- Introduction

  - motivation
  - review of other researches 
  - open questions in the domain
  - short summary of your proposed approach
