# Method


### Super Classs Definition

```python
import numpy as np 
from abc import ABC, abstractmethod

# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    
class LinearModel(BaseModel):
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):
        """
            train linear model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """        
        pass
    
    @abstractmethod
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """        
        pass 
```

### Least Squares

```python
# Linear Regression Class for least squares
class LinearRegress(LinearModel): 
    """ 
        LinearRegress class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        LinearModel.__init__(self)
        
    # train lease-squares model
    def train(self, X, T):
        X1 = self.add_ones(X)
        self.w = np.linalg.solve(X1.T @ X1, X1.T @ T)
    
    # apply the learned model to data X
    def use(self, X):
        X1 = self.add_ones(X)
        return X1 @ self.w
```

### Least Mean Squares

```python
import collections # for checking iterable instance

# LMS class 
class LMS(LinearModel):
    """
        Lease Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, alpha):
        LinearModel.__init__(self)
        self.alpha = alpha
        
    def train_step(self, x, t):
        x.shape=(1,x.shape[0]); #1,3
        x1 = self.add_ones(x)
        if(self.w is None):
            ##self.w=np.zeros((1,x1.shape[1])) # initialized w to zeros(1,3) 
            self.w = np.random.rand(1,x1.shape[1])
        y = np.inner(self.w , x1)
        self.w =self.w - self.alpha * (y - t) * x1
    
    # batch training by using train_step function
    def train(self, X, T):
        for x, t in zip(X, T):
            self.train_step(x, t)
            
    
    # apply the current model to data X
    def use(self, X):
        X1 = self.add_ones(X)
        return X1 @ self.w.T
        
```

```python
from sklearn import preprocessing
instances=10000 # Reduce this value for the better graphical visualization
X = np.array(df.loc[:instances, ['sqft_living']])
X= X.reshape((-1, 1))
#print(X)
print(X.shape)
#print(X[21])
T = np.array(df.loc[:instances, 'price'])
T= T.reshape((-1, 1))
#print(T.shape)
X_scaled = preprocessing.scale(X)
T_scaled = preprocessing.scale(T)
```
